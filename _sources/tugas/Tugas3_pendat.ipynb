Langkah 1 : Mengimport Perpustakaan

langkah pertama mengimpor pustaka yaitu NumPy, Pandas, dan Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Langkah 2 : Mengimpor kumpulan data

Pada tahap ini, kita akan mengimpor dataset Bunga Iris dari GithubGist dengan nama file IrisDataset.csv dan menyimpan hasilnya dalam variabel dataset. Selanjutnya, kita akan menetapkan 4 variabel independen dari dataset ke variabel X dan variabel dependen 'species' ke variabel Y. Untuk melihat data awal dari dataset yang telah diimpor, kita akan menggunakan fungsi head() dan menampilkan 5 baris pertama dari dataset tersebut. Hal ini akan membantu kita memahami data awal dari dataset yang akan digunakan.

```python
dataset = pd.read_csv('https://gist.githubusercontent.com/hamam14/d45a4b5c468a1265f7d8f109e62b8ab8/raw/20e78bf42c94eb378017a26599ba243d0ae108dd/IrisDataset.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

Langkah 3 : Memisahkan dataset menjadi set Pelatihan dan set Uji

Langkah selanjutnya setelah mengimpor dataset Bunga Iris adalah membaginya menjadi kumpulan pelatihan dan pengujian. Untuk memastikan pembagian data yang acak, kita menggunakan parameter test_size=0.2 yang akan membagi dataset menjadi 20% sebagai data pengujian dan 80% sebagai data pelatihan untuk melatih model klasifikasi Naive Bayes. Dengan memisahkan dataset menjadi data pelatihan dan pengujian, kita dapat menghindari terjadinya overfitting dan menguji kinerja model pada data yang belum pernah dilihat sebelumnya. Dataset Bunga Iris terdiri dari 150 baris data dengan masing-masing 50 baris dari 3 kelas yang diberikan secara berurutan.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

Langkah 4 : Penskalaan Fitur

Pada tahap ini, kita akan melakukan penskalaan fitur pada nilai variabel X_train dan X_test dengan menggunakan metode seperti StandardScaler atau MinMaxScaler. Dengan melakukan penskalaan fitur, kita dapat memperkecil rentang nilai dari setiap fitur pada dataset sehingga dapat mempercepat proses perhitungan pada model klasifikasi Naive Bayes. Penskalaan fitur ini dapat meningkatkan kecepatan program dan mengoptimalkan kinerja model klasifikasi Naive Bayes pada dataset yang lebih kecil ini.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Langkah 5 : Latih model Naive Bayes Classification pada Set Pelatihan

Pada tahap ini, kita akan menggunakan kelas GaussianNB dari pustaka sklearn.naive_bayes untuk melakukan klasifikasi pada dataset yang telah dibagi menjadi data pelatihan dan pengujian serta telah dilakukan penskalaan fitur. Model yang digunakan adalah model Gaussian dan objek kelas GaussianNB diinisiasi pada variabel classifier. Selanjutnya, nilai X_train dan y_train disesuaikan (fit) pada objek kelas GaussianNB untuk tujuan pelatihan. Dengan melatih model klasifikasi pada dataset pelatihan, kita dapat menggunakannya untuk memprediksi nilai kelas pada dataset pengujian (test set). Terdapat beberapa jenis model klasifikasi lainnya yang dapat digunakan seperti Bernoulli, Categorical, dan Multinomial.

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

Langkah 6 : Memprediksi hasil Test set

Pada tahap ini, setelah model GaussianNB dilatih pada dataset pelatihan, kita dapat menggunakan metode classifier.predict() untuk melakukan prediksi nilai kelas pada dataset pengujian (test set) dan hasil prediksi tersebut akan disimpan pada variabel y_pred. Dalam melakukan prediksi, model akan memproses nilai fitur dari setiap sampel pada test set dan menghasilkan nilai prediksi kelas berdasarkan model Gaussian yang telah dilatih pada dataset pelatihan. Dengan menggunakan nilai prediksi pada variabel y_pred, kita dapat mengevaluasi kinerja model klasifikasi pada dataset pengujian.

```python
y_pred = classifier.predict(X_test) 
y_pred
```

Langkah 7 : Matriks Kebingungan dan Akurasi

Setelah melakukan prediksi pada test set, kita dapat mengevaluasi kinerja model klasifikasi dengan menghitung akurasi model dan membuat confusion matrix. Confusion matrix adalah tabel yang menunjukkan jumlah prediksi yang benar dan salah pada suatu masalah klasifikasi ketika nilai sebenarnya dari test set diketahui. Tabel ini memiliki empat sel yaitu True Positive (TP), False Positive (FP), True Negative (TN), dan False Negative (FN). TP adalah jumlah prediksi yang benar dari kelas positif, FP adalah jumlah prediksi yang salah dari kelas negatif, TN adalah jumlah prediksi yang benar dari kelas negatif, dan FN adalah jumlah prediksi yang salah dari kelas positif.

Setelah prediksi dilakukan pada dataset pengujian, akurasi model klasifikasi GaussianNB dihitung dan ditampilkan. Setelah itu, confusion matrix dihasilkan dan ditampilkan dalam bentuk plot untuk memvisualisasikan kinerja model dengan lebih jelas. Melalui plot confusion matrix, kita dapat melihat seberapa baik model dapat mengklasifikasikan data dengan benar pada setiap kelas.

Hasil Akurasi :

nilai akurasi yang diperoleh dari model klasifikasi GaussianNB yaitu 0.9333 dan confusion matrix yang telah dihitung dan ditampilkan dalam bentuk array. Confusion matrix tersebut menunjukkan jumlah prediksi yang benar dan salah pada setiap kelas, yaitu 9 prediksi benar pada kelas 0, 11 prediksi benar pada kelas 1, dan 8 prediksi benar pada kelas 2. Sedangkan pada kelas 1, terdapat 2 prediksi salah.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

Langkah 8 : Membandingkan Nilai Riil dengan Nilai Prediksi

Pada tahap ini, dibuat sebuah Pandas DataFrame untuk membandingkan nilai klasifikasi dari test set asli (y_test) dan hasil prediksi (y_pred). Tahap ini dapat memberikan gambaran yang lebih jelas tentang kinerja model meskipun tidak se-informatif confusion matrix. Terdapat satu prediksi yang salah, di mana bunga iris yang seharusnya diklasifikasikan sebagai virginica telah diprediksi sebagai versicolor. Secara keseluruhan, model klasifikasi Naive Bayes memiliki akurasi yang sangat baik, dengan akurasi mencapai 96,67%.

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```
