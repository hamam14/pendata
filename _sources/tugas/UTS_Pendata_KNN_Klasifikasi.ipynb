Pada dasarnya, KNN merupakan algoritma supervised machine learning yang digunakan untuk mencari pola pada data yang ada dengan tujuan untuk membuat prediksi. Prediksi tersebut dapat berupa memprediksi harga mobil atau mengidentifikasi pelanggan yang akan merespons kampanye pemasaran email baru dengan baik.

Konsep KNN yang sederhana adalah dengan mengidentifikasi titik data yang memiliki fitur yang mirip dengan titik data yang ingin diprediksi, kemudian menuliskannya berdasarkan kemiripan (similarity), dan memilih k jumlah akhir dari daftar tersebut. Prediksi kemudian dapat dilakukan dengan mengambil rata-rata atau mean dari fitur dari titik akhir.

Dalam implementasi algoritma KNN, kita menggunakan metrik kemiripan untuk menghitung kemiripan antara data. Metrik kemiripan digunakan untuk membandingkan satu set fitur numerik yang tetap antara dua pengamatan.

1. Import library yang diperlukan

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.preprocessing import LabelEncoder
```

2. Mengambil data yang akan kita analisa

```python
df = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')

jumlah_baris = df.shape[0]
df.head(jumlah_baris)
```

3. Mengecek nilai yang hilang/kosong

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

4. Ubah nilai yang terdapat tanda tanya (?) menjadi nilai kosong

```python
df_copy = df_copy.replace('?',np.nan)
```

5. Cek lagi berapa nilai yang kosong pada data

```python
df_copy.isnull().sum()
```

6. Untuk melakukan normalisasi data, salah satu metode yang dapat digunakan adalah metode min-max. Metode ini mengubah setiap nilai pada kolom tertentu ke dalam rentang antara 0 hingga 1 dengan menggunakan rumus (x - min) / (max - min). Terdapat juga beberapa metode normalisasi lain seperti simple feature scaling atau z-score, namun perlu diingat bahwa kolom target (dalam kasus ini, kolom species) tidak perlu dinormalisasi.

```python
target_column = df_copy['target']
df_copy = df_copy[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].copy()
df_copy = (df_copy - df_copy.min())/(df_copy.max() - df_copy.min())
df_copy['target'] = target_column.copy()
df_copy.head(jumlah_baris)
```

7. Langkah terakhirnya adalah membuat objek knn dari kelas KNeighborsRegressor, yang menerapkan algoritma K-Nearest Neighbors untuk melakukan regresi. Selanjutnya, kita melakukan randomisasi indeks dari dataset dengan menggunakan np.random.permutation() dan seed 1, sehingga dapat memastikan bahwa hasil yang dihasilkan konsisten.

  > selanjutnya adalah membagi dataset menjadi data training sebesar 75% dan data testing sebesar 25% menggunakan variabel train_index. Data training akan disimpan di dalam variabel training_data, sementara data testing akan disimpan di dalam variabel testing_data.

  > Setelah melakukan pembagian dataset, kita melakukan pelatihan model dengan menggunakan data training. Hal ini dilakukan dengan memanggil method fit() pada objek knn, dengan argumen input berupa fitur-fitur yang digunakan (yaitu sepal_length, sepal_width, petal_length, dan petal_width) dan label yaitu species.

  > Setelah pelatihan model selesai, kita melakukan prediksi pada data testing dengan memanggil method predict() pada objek knn. Input yang diberikan pada method predict() adalah fitur-fitur dari data testing yang ingin diprediksi.

  > Setelah melakukan prediksi, kita menghitung akurasi model dengan mengukur berapa persen data testing yang berhasil diprediksi dengan menggunakan threshold default 0.5.

  > Terakhir, mencetak hasil akurasi model.

```python
knn = KNeighborsRegressor()

# before we split, make sure you randomize the dataset
np.random.seed(1)
index_shuffled = np.random.permutation(df_copy.index)
shuffled_car_data = df_copy.loc[index_shuffled]

# We will use 75% of the data for training and the remaining 25% for testing
train_index = int(shuffled_car_data.shape[0] * 0.75)
training_data = shuffled_car_data.iloc[0:train_index]
testing_data = shuffled_car_data.iloc[train_index:]

# fit the model with the training data
knn.fit(training_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']], training_data['target'])

# predict the target of the testing data
prediction = knn.predict(testing_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']])

# calculate the accuracy of the model
accuracy = np.mean(np.abs(prediction - testing_data['target']) <= 0.5)

print(f"Akurasi model KNN: {accuracy:.2%}")
```

KESIMPULAN :

> Dari hasil yang telah diperoleh, diketahui bahwa tingkat akurasi model KNN pada dataset iris dengan menggunakan beberapa fitur ('sepal_length', 'sepal_width', 'petal_length', 'petal_width') adalah sebesar 92.11%.
