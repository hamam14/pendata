```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances
data = pd.read_csv('https://raw.githubusercontent.com/hamam14/pendata/data/data_pekerjaan.csv')
data
```

```python
X = data[['nik', 'jenis kelamin', 'jabatan', 'pns']].copy()
X['nik'] = pd.Categorical(X['nik']).codes
X['jenis kelamin'] = pd.Categorical(X['jenis kelamin']).codes
X['jabatan'] = pd.Categorical(X['jabatan']).codes
X['pns'] = pd.Categorical(X['pns'], categories=['staf', 'bendahara', 'camat'], ordered=True).codes
```

```python
dist_matrix = pairwise_distances(X, metric='manhattan') / max(pairwise_distances(X, metric='manhattan').flatten())
```
